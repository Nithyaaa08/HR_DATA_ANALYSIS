```python
# Convert object columns to category type
categorical_cols = ['Attrition', 'BusinessTravel', 'Department', 'Gender',
                    'EducationField', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']

for col in categorical_cols:
    df[col] = df[col].astype('category')

# Check data types again
df.dtypes
```

<p style="text-align:justify; font-size:20px; color:black;">
Several columns containing categorical text data (e.g., <strong>Attrition</strong>, <strong>Gender</strong>, <strong>Department</strong>) were originally of <strong>object</strong> type. These columns were converted to <strong>category</strong> type using <strong>astype('category')</strong> to reduce memory usage and improve performance for aggregation, filtering, and visualization tasks. Converting object columns to category type also makes the dataset more efficient for analytics.
</p>

```python
# Convert Attrition from Yes/No to numeric
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
df[['Attrition', 'Attrition_Flag']].head(10)
```

<p style="text-align:justify; font-size:20px; color:black;">
The <strong>Attrition</strong> column in the dataset originally contained categorical values: "Yes" for employees who left the company and "No" for those who stayed. To enable numerical analysis and simplify computations, this column was converted into a new binary column named <strong>Attrition_Flag</strong>, where Yes is represented as 1 and No as 0. This transformation allows easy calculation of attrition rates, statistical analysis, and aggregation by other features, such as experience or income. It also prepares the data for any predictive modeling or visualization that requires numeric input.
</p>

```python
# Categorize employees based on years at company
df['ExperienceCategory'] = pd.cut(df['YearsAtCompany'],bins=[0, 2, 6, df['YearsAtCompany'].max()],labels=['New', 'Mid-Level', 'Senior'])
df[['YearsAtCompany', 'ExperienceCategory']].head(10)
```

<p style="text-align:justify; font-size:20px; color:black;">
Employees were categorized based on the number of years they have spent in the company, using the <strong>YearsAtCompany</strong> variable. A new column <strong>ExperienceCategory</strong> was created with three categories: <strong>New</strong> (0–2 years), <strong>Mid-Level</strong> (3–6 years), and <strong>Senior</strong> (7 years and above). This derived column helps analyze patterns of attrition and performance according to experience levels. For example, it can show whether newer employees are more likely to leave or if senior employees have higher satisfaction, enabling managers to take targeted retention measures.
</p>

```python
# Create salary bands
df['IncomeCategory'] = pd.cut(df['MonthlyIncome'],bins=[0, 5000, 10000, df['MonthlyIncome'].max()],labels=['Low Income', 'Medium Income', 'High Income'])
df[['MonthlyIncome', 'IncomeCategory']].head(10)
```

<p style="text-align:justify; font-size:20px; color:black;">
The <strong>MonthlyIncome</strong> column was transformed into a categorical variable named <strong>IncomeCategory</strong> to create clear salary bands. Employees were classified into <strong>Low Income</strong> (≤5,000), <strong>Medium Income</strong> (5,001–10,000), and <strong>High Income</strong> (>10,000). This grouping allows a better understanding of how income levels relate to attrition and job satisfaction. It provides insight into whether employees in specific income bands are more likely to leave or perform differently, which can guide compensation planning and retention strategies.
</p>

```python
# Convert work-life balance scale to labels
df['WorkLifeBalance_Label'] = df['WorkLifeBalance'].map({1: 'Poor',2: 'Average',3: 'Good',4: 'Excellent'})
df[['WorkLifeBalance', 'WorkLifeBalance_Label']].head(10)
```

<p style="text-align:justify; font-size:20px; color:black;">
The numeric <strong>WorkLifeBalance</strong> column, which had values from 1 to 4, was converted into a more interpretable categorical column called <strong>WorkLifeBalance_Label</strong>. The mapping is: 1 → <strong>Poor</strong>, 2 → <strong>Average</strong>, 3 → <strong>Good</strong>, 4 → <strong>Excellent</strong>. This transformation makes it easier to analyze work-life balance patterns and their impact on employee attrition or performance. By using meaningful labels instead of numbers, the dataset becomes more readable for reporting and visualization, enabling managers to identify areas where work-life balance improvements may reduce turnover.
</p>

<h3 style="font-size:25px; color:black; font-weight:bold; text-align:left;">
3.3 Multivariate Analysis
</h3>

<p style="text-align:justify; font-size:20px; color:black;">
Multivariate analysis studies the relationships between <strong>more than two variables</strong> at the same time. This helps uncover complex patterns, such as how factors like <strong>department, income, experience, and work-life balance</strong> together influence employee attrition or performance. It provides deeper insights than univariate or bivariate analysis alone.
</p>

```python
# Create the pivot table AND store it in a variable
pivot = pd.pivot_table(
    df,
    values='Attrition_Flag',
    index='Department',
    columns='WorkLifeBalance_Label',
    aggfunc='mean',
    observed=True
)

# Convert attrition rate to percentage
pivot_percent = pivot * 100

pivot_percent
```

<p style="text-align:justify; font-size:20px; color:black;">
A pivot table was created to study attrition rates across <strong>departments</strong> and <strong>work-life balance levels</strong> simultaneously. This multivariate analysis helps identify risky combinations where attrition is high.
</p>

```python
import matplotlib.pyplot as plt
```

```python
# Overall attrition count
df['Attrition'].value_counts().plot(kind='bar', color='purple')

plt.title('Overall Employee Attrition Count')
plt.xlabel('Attrition Status')
plt.ylabel('Number of Employees')
plt.show()
```

<p style="text-align:justify; font-size:20px; color:black;">

```python
# Attrition percentage
df['Attrition'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title('Employee Attrition Percentage')
plt.ylabel('')
plt.show()
```

<p style="text-align:justify; font-size:20px; color:black;">

```python
# Attrition rate by department
df.groupby('Department', observed=True)['Attrition_Flag'].mean().plot(kind='bar', color='skyblue')

plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Attrition by work-life balance
df.groupby('WorkLifeBalance_Label', observed=True)['Attrition_Flag'].mean().plot(kind='bar',color='pink')

plt.title('Attrition Rate by Work-Life Balance')
plt.xlabel('Work-Life Balance')
plt.ylabel('Attrition Rate')
plt.show()
```
