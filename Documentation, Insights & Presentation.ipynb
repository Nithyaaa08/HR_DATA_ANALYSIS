<h1 style="text-align:center; font-weight:bold; font-size:40px; color:black;">
HR Analytics: Employee Attrition Analysis Using Exploratory Data Analysis
</h1>

<h2 style="font-size:30px; font-weight:bold; color:black;">
1. Project Objective
</h2>

<p style="text-align:justify; font-size:20px; color:black;">
Employee attrition is a major concern for organizations as it directly impacts productivity, recruitment costs, and workforce stability.
The objective of this project is to analyze employee data to understand the key factors contributing to attrition.
</p>

<p style="font-size:20px; color:black;">
This project aims to:
</p>

<ul style="font-size:20px; color:black;">
  <li>Study employee demographics, job-related factors, and satisfaction levels</li>
  <li>Identify patterns and trends associated with employee attrition</li>
  <li>Use exploratory data analysis (EDA) and visualizations to derive insights</li>
  <li>Provide meaningful business recommendations to help reduce attrition</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
The analysis is performed using Python libraries such as <strong>Pandas</strong>, <strong>Matplotlib</strong>, and <strong>Seaborn</strong>.
</p>

<h2 style="font-size:30px; font-weight:bold; color:black;">
2. Data Understanding and Preprocessing
</h2>

<p style="text-align:justify; font-size:20px; color:black;">
The dataset contains employee-related information such as age, department, job role, income, experience, job satisfaction, work-life balance, and attrition status.
</p>

<h3 style="font-size:25px; color:black;">
2.1 Initial Data Inspection
</h3>

<p style="font-size:20px; color:black;">
The dataset was explored using:
</p>

<ul style="font-size:20px; color:black;">
  <li><code>df.head()</code> to understand the structure of the dataset</li>
  <li><code>df.info()</code> to check data types and identify missing values</li>
  <li><code>df.describe()</code> to obtain statistical summaries of numeric variables</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
These steps helped in understanding the nature of the variables and their distributions, and guided further data preprocessing decisions.
</p>

<h3 style="font-size:25px; color:black;">
2.2 Missing Value Analysis
</h3>

<p style="text-align:justify; font-size:20px; color:black;">
A missing value analysis was performed using the <code>df.isnull().sum()</code> function.
The results showed that <strong>all columns contained zero missing values</strong>.
</p>

<p style="text-align:justify; font-size:20px; color:black;">
Since no missing data was present, <strong>no imputation or data cleaning for missing values was required</strong>.
This confirms that the dataset was complete and suitable for further analysis without unnecessary preprocessing.
</p>

<h3 style="font-size:25px; color:black;">
2.3 Duplicate Value Check
</h3>

<p style="text-align:justify; font-size:20px; color:black;">
As part of data preprocessing, the dataset was checked for duplicate records using the <code>df.duplicated().sum()</code> function.
The result showed that there were <strong>no duplicate rows</strong> in the dataset.
</p>

<p style="text-align:justify; font-size:20px; color:black;">
This confirms that each employee record is unique, and no duplicate removal was required before proceeding with analysis.
</p>

<h3 style="font-size:24px; color:black;">
2.4 Data Type Optimization
</h3>

<p style="text-align:justify; font-size:20px; color:black;">
The <code>df.dtypes</code> function was used to inspect the data types of each column.
Categorical variables such as <strong>Attrition, Department, Job Role, Gender,</strong> and <strong>OverTime</strong> were converted to the <code>category</code> data type.
</p>

<p style="text-align:justify; font-size:20px; color:black;">
This step improves memory efficiency and ensures appropriate handling of categorical variables during analysis and visualization.
</p>

<h3 style="font-size:25px; color:black;">
2.5 Derived Columns
</h3>

<p style="text-align:justify; font-size:20px; color:black;">
To enhance analysis and interpretation, additional derived columns were created:
</p>

<ul style="font-size:20px; color:black;">
  <li><strong>Attrition_Flag</strong>: Converted Attrition from Yes/No to numeric (1/0)</li>
  <li><strong>ExperienceCategory</strong>: Categorized employees as New, Mid-Level, or Senior based on YearsAtCompany</li>
  <li><strong>IncomeCategory</strong>: Grouped MonthlyIncome into Low, Medium, and High income bands</li>
  <li><strong>WorkLifeBalance_Label</strong>: Converted numeric work-life balance scores into meaningful labels</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
These derived columns helped simplify comparisons and improve the clarity of visualizations.
</p>

<h2 style="font-size:30px; font-weight:bold; color:black;">
3. Exploratory Data Analysis (EDA)
</h2>

<p style="text-align:justify; font-size:20px; color:black;">
The purpose of Exploratory Data Analysis (EDA) is to understand the dataset in depth by identifying patterns, trends, and relationships among variables.
In this step, statistical summaries and grouping techniques are used to analyze employee attrition and performance from different perspectives.
</p>

<p style="font-size:20px; color:black;">
EDA helps in answering important business questions such as:
</p>

<ul style="font-size:20px; color:black;">
  <li>Which employees are more likely to leave?</li>
  <li>How do factors like department, income, experience, and work-life balance affect attrition?</li>
  <li>What relationships exist among key numeric variables?</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
This step forms the foundation for meaningful insights and data-driven decision-making.
</p>

<h3 style="font-size:24px; color:black;">
3.1 Univariate Analysis
</h3>

<p style="font-size:20px; color:black;">
Analysis of individual variables
</p>

<p style="font-size:22px; color:black;">
<strong>1. Attrition Distribution</strong>
</p>

<p style="text-align:justify; font-size:20px; color:black;">
<strong>Observation:</strong> Out of 1,470 employees,
</p>

<ul style="font-size:20px; color:black;">
  <li>1,233 employees stayed with the organization</li>
  <li>237 employees left the organization</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
This indicates that while the majority of employees are retained, there is a notable attrition rate that requires further investigation.
</p>

<p style="text-align:justify; font-size:20px; color:black;">
<strong>Interpretation:</strong> The imbalance between employees who stayed and those who left highlights the importance of understanding the factors contributing to attrition.
This variable becomes the key target variable for further analysis.
</p>

<p style="font-size:22px; color:black;">
<strong>2. Age Summary Statistics</strong>
</p>

<p style="font-size:20px; color:black;">
<strong>Key Statistics:</strong>
</p>

<ul style="font-size:20px; color:black;">
  <li>Average age: ~37 years</li>
  <li>Minimum age: 18 years</li>
  <li>Maximum age: 60 years</li>
  <li>50% of employees fall between 30 and 43 years</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
<strong>Interpretation:</strong> The workforce is primarily composed of mid-career professionals, with a balanced age distribution.
This suggests that attrition is not limited to a very young or very old workforce and should be examined across different experience levels.
</p>

<h3 style="font-size:25px; color:black;">
3.2 Bivariate Analysis
</h3>

<p style="font-size:20px; color:black;">
Relationship between two variables
</p>

<p style="font-size:22px; color:black;">
<strong>1. Attrition by Department</strong>
</p>

<p style="font-size:20px; color:black;">
<strong>Findings:</strong>
</p>

<ul style="font-size:20px; color:black;">
  <li>Sales: ~20.6% attrition (highest)</li>
  <li>Human Resources: ~19.0%</li>
  <li>Research &amp; Development: ~13.8% (lowest)</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
<strong>Interpretation:</strong> Sales and Human Resources departments experience significantly higher attrition compared to Research &amp; Development.
This may be due to job pressure, performance targets, or limited growth opportunities.
Research &amp; Development appears more stable, possibly due to specialized roles and clearer career paths.
</p>

<p style="font-size:22px; color:black;">
<strong>2. Attrition by Work-Life Balance</strong>
</p>

<p style="font-size:20px; color:black;">
<strong>Findings:</strong>
</p>

<ul style="font-size:20px; color:black;">
  <li>Poor work-life balance: ~31.3% attrition (highest)</li>
  <li>Average work-life balance: ~16.9%</li>
  <li>Excellent work-life balance: ~17.6%</li>
  <li>Good work-life balance: ~14.2% (lowest)</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
<strong>Interpretation:</strong> Employees reporting poor work-life balance are more than twice as likely to leave compared to those with good balance.
This clearly indicates that work-life balance is a critical driver of attrition.
</p>
<p style="font-size:22px; color:black;">
<strong>3. Attrition by Income Category</strong>
</p>

<p style="font-size:20px; color:black;">
<strong>Findings:</strong>
</p>

<ul style="font-size:20px; color:black;">
  <li>Low Income: ~21.8% attrition</li>
  <li>Medium Income: ~11.1%</li>
  <li>High Income: ~8.9%</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
<strong>Interpretation:</strong> Attrition decreases as income increases.
Employees in lower income categories are significantly more likely to leave, suggesting that compensation plays a strong role in employee retention.
</p>

<h3 style="font-size:25px; color:black;">
3.3 Multivariate Analysis
</h3>

<p style="font-size:20px; color:black;">
More than two variables together
</p>

<p style="font-size:22px; color:black;">
<strong>1. Attrition by Department &amp; Work-Life Balance (Pivot Table)</strong>
</p>

<p style="font-size:20px; color:black;">
<strong>Key Patterns Observed:</strong>
</p>

<ul style="font-size:20px; color:black;">
  <li>Sales + Poor Work-Life Balance: ~37.5% attrition (highest risk group)</li>
  <li>R&amp;D + Poor Work-Life Balance: ~31.7%</li>
  <li>HR + Average Work-Life Balance: ~28.6%</li>
  <li>Departments with good or excellent work-life balance consistently show lower attrition</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
<strong>Interpretation:</strong> Attrition is not driven by a single factor alone.
Employees working in high-pressure departments combined with poor work-life balance are at the highest risk of leaving.
This insight is valuable for designing targeted retention strategies rather than applying one-size-fits-all solutions.
</p>

<h3 style="font-size:25px; color:black;">
3.4 Correlation Analysis
</h3>

<p style="font-size:20px; color:black;">
Relationships among numeric variables
</p>

<p style="font-size:22px; color:black;">
<strong>1. Key Correlation Insights</strong>
</p>

<p style="font-size:20px; color:black;">
<strong>Strong Positive Correlations:</strong>
</p>

<ul style="font-size:20px; color:black;">
  <li>MonthlyIncome ↔ JobLevel (very strong)</li>
  <li>TotalWorkingYears ↔ JobLevel</li>
  <li>YearsAtCompany ↔ YearsWithCurrManager</li>
  <li>YearsAtCompany ↔ YearsInCurrentRole</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
These relationships indicate natural career progression — as employees gain experience and tenure, their job level, income, and managerial stability increase.
</p>

<p style="font-size:20px; color:black;">
<strong>Moderate Relationships:</strong>
</p>

<ul style="font-size:20px; color:black;">
  <li>Age is moderately correlated with experience and job level</li>
  <li>Employees with longer tenure tend to have longer relationships with managers and roles</li>
</ul>

<p style="font-size:20px; color:black;">
<strong>Low or Negligible Correlations:</strong>
</p>

<ul style="font-size:20px; color:black;">
  <li>DailyRate, HourlyRate, and MonthlyRate show weak correlations with most variables</li>
  <li>PerformanceRating shows little variation, suggesting limited influence on attrition</li>
</ul>

<p style="text-align:justify; font-size:20px; color:black;">
<strong>Interpretation:</strong> Experience, income, and job stability tend to move together and are likely to contribute to employee retention.
Employees with shorter tenure, lower job levels, and lower income appear more vulnerable to attrition.
</p>
